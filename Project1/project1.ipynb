```python
# Finding Lane Lines on the Road
# Kwanghyun JUNG (ihangulo@gmail.com)
# 2 Nov 2016
# Do relevant imports
# github.com/udacity/CarND-LaneLines-P1
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
import numpy as np
import cv2
import math
import os

# to show interactive result
%matplotlib inline 


### help functions
    
def grayscale(img):
    """Applies the Grayscale transform
    This will return an image with only one color channel
    but NOTE: to see the returned image as grayscale
    you should call plt.imshow(gray, cmap='gray')"""
    return cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
    
def canny(img, low_threshold, high_threshold):
    """Applies the Canny transform"""
    return cv2.Canny(img, low_threshold, high_threshold)

def gaussian_blur(img, kernel_size):
    """Applies a Gaussian Noise kernel"""
    return cv2.GaussianBlur(img, (kernel_size, kernel_size), 0)

def region_of_interest(img, vertices):
    """
    Applies an image mask.
    
    Only keeps the region of the image defined by the polygon
    formed from `vertices`. The rest of the image is set to black.
    """
    #defining a blank mask to start with
    mask = np.zeros_like(img)   
    
    #defining a 3 channel or 1 channel color to fill the mask with depending on the input image
    if len(img.shape) > 2:
        channel_count = img.shape[2]  # i.e. 3 or 4 depending on your image
        ignore_mask_color = (255,) * channel_count
    else:
        ignore_mask_color = 255
        
    #filling pixels inside the polygon defined by "vertices" with the fill color    
    cv2.fillPoly(mask, vertices, ignore_mask_color)
    
    #returning the image only where mask pixels are nonzero
    masked_image = cv2.bitwise_and(img, mask)
    return masked_image
```

```python
def draw_lines2(img, lines, color=[255,0,0], thickness=10, center_x = 480, center_y =325) :
    """
    NOTE: this is the function you might want to use as a starting point once you want to 
    average/extrapolate the line segments you detect to map out the full
    extent of the lane (going from the result shown in raw-lines-example.mp4
    to that shown in P1_example.mp4).  
    
    Think about things like separating line segments by their 
    slope ((y2-y1)/(x2-x1)) to decide which segments are part of the left
    line vs. the right line.  Then, you can average the position of each of 
    the lines and extrapolate to the top and bottom of the lane.
    """
    
    
    slope_left = []
    slope_right = []
        #get coord which have min&max y-intercept  each side
        #y=mx+b --> b= y-mx
    left_first = True
    right_first= True
    
    
    screen_height= img.shape[0]
    screen_width = img.shape[1]
    
    for line in lines :
        for x1, y1, x2, y2 in line :
            slope = ((y2-y1)/(x2-x1))
            tempB = y1 - slope * x1 # get y-intercept (y-y1)= m(x-x1) -> y=mx -mx1 + y1 // tempB는 y절편 값이다.
            
         
            if  (slope > 0.5 and slope < 5.7 and x1>= center_x  and tempB < 50)     : #right side lane 오른쪽 선 80도이하
                # the slope is tan(x) is bigger than 30(degree) smaller than 80
                # Y-intercept must smaller than 50 (in fact 0, but there is so many exceptions)
                #만약 y절편이 0보다 크면 수평에 가까운 것이므로 무시한다. 즉 0보다 작아야 한다. 하지만 오차를 둔다.
                #is really right side?
                
                slope_right.append(slope)
         

                if(right_first) : #if first data 만약 처음 데이터이면
                    
                    right_max_b_value = tempB 
                    right_max_b = (x1, y1)  
                    right_min_b_value = tempB # smallest y-intercept 최소 y절편 값
                    right_min_b = (x1, y1)    # biggest y-intercept
                    right_min_y = (x1, y1) 
                    
                    right_first=False
                    
                else :
                    if (tempB > right_max_b_value) :
                        right_max_b_value = tempB # b최대 y절편 값
                        right_max_b = (x1, y1)
                
                    if(tempB < right_min_b_value) :
                        right_min_b_value = tempB #최소 y절편 값
                        right_min_b = (x1, y1)
                        
                    if(y1 < right_min_y[1]) :
                        right_min_y = (x1, y1)
                    
                    
            elif (slope < -0.5 and  slope > -5.7 and x1 <= center_x and tempB > screen_height) : # left_side lane
                # the slope is tan(x) is bigger than 30(degree) smaller than 80
                # Y-intercept must bigger than screen_height
                #left side 왼쪽 선 y절편이 최소한 스크린보다는 커야 한다(사다리꼴)
                # 만약 y 절편이 현재 스크린 세로 사이즈 보다 작으면 무시해야 한다. (곡선일 가능성) 하지만 좀 오차를 둔다.
                slope_left.append(slope)  
                
                if(left_first) :
                    
                    left_max_b_value = tempB
                    left_max_b = (x1, y1)
                    left_min_b_value = tempB
                    left_min_b = (x1, y1)
                    left_min_y = (x1,y1)
                    
                    left_first=False

                else :
                    
                    if (tempB > left_max_b_value) :
                        left_max_b_value = tempB
                        left_max_b = (x1, y1)
                   
                    if(tempB < left_min_b_value) :
                        left_min_b_value = tempB
                        left_min_b = (x1, y1)
                  

                    #find min y 
                    
                    if(y1 < left_min_y[1]) :
                        left_min_y = (x1, y1)

                    
                
    # ok. everage slope get. the line intersect bottom line(= y pos = image end)
    if(len(slope_left)>0) :
     
        
        left_slope_avg =  sum(slope_left) / len(slope_left)

        left_center = ( int ((left_max_b[0] + left_min_b[0]) /2 ), int( (left_max_b[1] + left_min_b[1]) /2))

    
        # right line is on the intermediate point max_b min b --> we can get lines
        # slope m and goes on (x1, y1) is y = m(x-x1)+y1 = mx- mx1 + y1  x= (y-y1+mx1)/m 
        # this two line ..지나간다. y=0 일때와 y=최대일때(img.shape[1])
        # if y=0
        # right_x1_Top = int(right_center[0] - right_center[1]/right_slope_avg)
        # when y= left_min_y_value
        
        # minValueVersion
        #left_x1_Top_min =  (int((left_min_y_value-left_center[1])/left_slope_avg +left_center[0] ), left_min_y_value)  
        
        #center value Version
        left_x1_Top =  (int((center_y-left_center[1])/left_slope_avg +left_center[0] ), center_y)  
        
        left_x2_Bottom = (int((screen_height-left_center[1])/left_slope_avg +left_center[0] ), screen_height)
        
        if(left_x1_Top[0] < center_x) :# check effective
            cv2.line(img, left_x1_Top , left_x2_Bottom, color, thickness)
        else : # then alternative (top of the spot) 
            cv2.line(img, left_min_y, left_x2_Bottom, color, thickness)
            
    
    if( len(slope_right)>0 ):

        right_slope_avg =  sum(slope_right) / len(slope_right)

        # the cetner point on the right line
        right_center = ( int ((right_max_b[0] + right_min_b[0]) /2 ), int( (right_max_b[1] + right_min_b[1]) /2))
    

        #min Value version
        #right_x1_Top_min =  (int((right_min_y_value-right_center[1])/right_slope_avg +right_center[0] ), right_min_y_value)  
        
        #center value version
        right_x1_Top =  (int((center_y-right_center[1])/right_slope_avg +right_center[0] ), center_y)  
        
        # when y = img.shape[1]
        right_x2_Bottom = (int((screen_height-right_center[1])/right_slope_avg +right_center[0] ), screen_height)
  
        if(right_x1_Top[0] > center_x) : # check effective
            cv2.line(img, right_x1_Top, right_x2_Bottom, color, thickness)            

        else : # then alternative (top of the spot)
            cv2.line(img, right_min_y, right_x2_Bottom, color, thickness)
        
```

```python
def draw_lines(img, lines, color=[255, 0, 0], thickness=4):
                
    """
    This function draws `lines` with `color` and `thickness`.    
    Lines are drawn on the image inplace (mutates the image).
    If you want to make the lines semi-transparent, think about combining
    this function with the weighted_img() function below
    """
    
    for line in lines:
        for x1,y1,x2,y2 in line:
            cv2.line(img, (x1, y1), (x2, y2), color, thickness)

def hough_lines(img, rho, theta, threshold, min_line_len, max_line_gap):
    """
    `img` should be the output of a Canny transform.
        
    Returns an image with hough lines drawn.
    """
    lines = cv2.HoughLinesP(img, rho, theta, threshold, np.array([]), minLineLength=min_line_len, maxLineGap=max_line_gap)
    line_img = np.zeros(img.shape, dtype=np.uint8)
    draw_lines(line_img, lines)
    return line_img

def hough_lines2(img, rho, theta, threshold, min_line_len, max_line_gap, centerX = 480 , centerY =325):
    """
    `img` should be the output of a Canny transform.
        
    Returns an image with hough lines drawn. // continuos line
    """
    lines = cv2.HoughLinesP(img, rho, theta, threshold, np.array([]), minLineLength=min_line_len, maxLineGap=max_line_gap)
    line_img = np.zeros(img.shape, dtype=np.uint8)
    draw_lines2(line_img, lines ,center_x=centerX, center_y = centerY) # draw solid line
    return line_img

# Python 3 has support for cool math symbols.

def weighted_img(img, initial_img, a=0.8, b=1., l=0.):
    """
    `img` is the output of the hough_lines(), An image with lines drawn on it.
    Should be a blank image (all black) with lines drawn on it.
    
    `initial_img` should be the image before any processing.
    
    The result image is computed as follows:
    initial_img * α + img * β + λ
    NOTE: initial_img and img must be the same shape!
    """
    return cv2.addWeighted(initial_img, a, img, b, l)


#main pipeline 

def process_image(image, mode=2):
    # NOTE: The output you return should be a color image (3 channel) for processing video below
    # TODO: put your pipeline here,
    # you should return the final output (image with lines are drawn on lanes)
     
        
    # make grey scale image
    gray = grayscale(image)

    # Define a kernel size and apply Gaussian smoothing
    kernel_size = 5
    blur_gray = gaussian_blur(gray, kernel_size)


    # Define our parameters for Canny and apply
    low_threshold = 30
    high_threshold = 150
    edges = canny(blur_gray, low_threshold, high_threshold)

    
    # This time we are defining a four sided polygon to mask
    imshape=image.shape
    vertices = np.array([[(160,imshape[0]),(430, 325), (525, 325), (imshape[1],imshape[0])]], dtype=np.int32)
    masked_edges = region_of_interest(edges, vertices)
 

    # Define the Hough transform parameters
    # Make a blank the same size as our image to draw on
    rho = 2
    theta = np.pi/180
    threshold = 20
    
    min_line_length = 30
    max_line_gap = 20
    
    
    if(mode==2) :
        hough_line_image = hough_lines2(masked_edges, rho, theta, threshold,min_line_length, max_line_gap)
    else :
        hough_line_image = hough_lines(masked_edges, rho, theta, threshold,min_line_length, max_line_gap)


    color_hough = np.dstack((hough_line_image, np.copy(hough_line_image)*0, np.copy(hough_line_image)*0)) 

    
    color_hough[:,:,1] = 0
    
 
    
    result=weighted_img( color_hough, image)
    return result
```

### about draw_lines2()

**- determin Left side lane or right side lane of my car **

it is very simple. we can examine slope (m) from two points.

~~~
    if (x1, y1) , (x2, y2) --> m = (y2-y1)/ (x2-x1) 
    
    if m<0 then left side / m > 0 then right side (it is differnt from math)

~~~
so we can make two list (left side/ right side) -> slope_left[] / slope_right[]



**- How to get average slope and one point on it.**

Let's think about only left side.
Then left side is like below picture.
~~~ 
    L1 L2
    / /
   / /
  / /
 / /
/ /
~~~

**L1** is the set of left-side line of left lane, **L2** is right side set.

(L1은 즉 차선의 왼쪽실선에 해당하는 라인에 해당하는 집합이라고 가정한다.

**Get the Point**

Standard solution is, draw horizontal line which is intersect two lines (L1, L2), and get the middle point of that intersection points. (It's very easy to draw line with two points) But this project, the line is so particled, so it's very hard to use this solution.

So, I use some trick.

(먼저  지나는 한 점을 찾아보자.
이는 실제로 수평선을 그어서 L1과 L2의 교점을 찾고 그 교점의 중점을 찾으면 된다. 하지만, L1과 L2는 파편화 되어 있어서 두 선을 지나는 수평선을 그으려면 많은 계산이 필요하다. 따라서, 여기서는 간단히 트릭을 찾아보기로 한다.)

First let's determine the line is L1 or L2. It's not easy. Because we have only know two end points of that. So it's very difficult to know the status of that line. But generally y-intercept of L1 is upper side of L2. So we can 
    the y coord of y-intercpet(L2) > y-intercept(L1)
    
(그런데, 일반적으로  L1은 L2보다 y절편이 위에 있다. 즉, L2의 y절편이 L1의 y절편보다 크다. y절편(L2) > y절편(L1))

So check the all y-intercept of all lines, get Max and Min value. Then we think that two points is positioned like below pictures. ( '*' is the spot which have min or max value.)

**Bottom End points (양 끝점)**

Bottom end point is so easy. Because it intersect with screen border(=screen heights).
(화면 하단의 끝점은 y좌표가 화면의 세로크기와 같은 점이므로 아래와 같이 쉽게 구할 수 있다. )

so..
~~~
   (eq.1)
   
    y = S(x - C1) + C2 --> x = (y-C2) / S  + C1

  Bottom end point:  P2(  (screen_height-C2) / S  + C1   , screen_height)
~~~  
  
**Top end point**

There can be exist various solutions. But I think two-step solution.

we can get "Vanishing point" if we have two intersecting lines. The intersection points is the vanishing point.

Let the two equations of  line

--------
~~~
    y = Ax + c
    y = Bx + d

    let's find intersection point.
    
    Ax + c = Bx + d

    x = (d-c)/ (A-B) 
    y = A( (d-c)/A-B ) + c 

    Intersection Point P = ( (d-c)/ (A-B) , A( (d-c)/A-B ) + c )

~~~
--------

But there is problem. If the average slope havs starange value, for example there is only one line with garbage value, we cannot check the effectiveness of that with ease. So it's slope is so strange but the average slope. So, in this case, if we get vanishing point, it is very right side of screen, beyond left lane. 

And this point is virtual point, so we cannot put this point into screen. if we draw from this point to bottom, and must delete the upper side of load.

So I make mind not use this point, instead use "top points" which is the center point of 'region of interest

~~~
    vertices = np.array([[(160,imshape[0]),(430, 325), (525, 325), (imshape[1],imshape[0])]], dtype=np.int32)

    masked_edges = region_of_interest(edges, vertices)
~~~
above, the 325 is the y position of uppper limit, so we can get easily x-position like below.

~~~

    y = S(x - C1) + C2
    let center_y (y-position of top masked_edges)
    
    x = (center_y-C2) / S  + C1

~~~
----
But this point can have same problem (invalid position) like vanising point. So, I use another "emergency value', the "Toppest point of lines". If there is top point which is get from line, then it has less error. So, draw line from bottom value to upper Toppest point at this situation.
